[Problem 1] Creating random numbers

```python
import numpy as np
```

```python
np.random.seed(42)

mean = (-3, 0)
covariance_matrix = np.asarray([[1.0, 0.8], [0.8, 1.0]])
arr = np.random.multivariate_normal(mean, covariance_matrix, size=500)
arr.shape
```

[Problem 2] Visualization by scatter plot

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(arr[:, 0], arr[:, 1])
plt.title("Scatter plot")
plt.show()
```

[Problem 3] Visualization by histogram

```python
plt.hist(arr, bins=100)
plt.xlim(-5, 2)
plt.show()
```

```python
plt.hist(arr[:, 0], bins=100)
plt.xlim(-5, 2)
# plt.xlabel("x")
# plt.ylabel("frequency")
plt.show()
```

```python
plt.hist(arr[:, 1], bins=100)
plt.xlim(-5, 2)
# plt.xlabel("x")
# plt.ylabel("frequency")
plt.show()
```

[Problem 4] Addition of data

```python
mean = (-3, 0)
covariance_matrix = np.asarray([[1.0, 0.8], [0.8, 1.0]])
arr1 = np.random.multivariate_normal(mean, covariance_matrix, size=500) 
```

```python
mean = (0, -3)
covariance_matrix = np.asarray([[1.0, 0.8], [0.8, 1.0]])
arr2 = np.random.multivariate_normal(mean, covariance_matrix, size=500) 
```

```python
plt.scatter(arr1[:, 0], arr1[:, 1], label="0")
plt.scatter(arr2[:, 0], arr2[:, 1], label="1")
plt.legend()
plt.show()
```

[Problem 5] Data combination

```python
arr_full = np.concatenate((arr1, arr2), axis=0)
arr_full.shape
```

```python
arr_full = np.vstack((arr1, arr2))
arr_full.shape
```

[Problem 6] Labeling

```python
label = np.zeros((500, 1))
label.shape
```

```python
arr1 = np.concatenate((arr1, label), axis=1)
arr1.shape
```

```python
label1 = np.ones((500, 1))
label1.shape
```

```python
arr2 = np.concatenate((arr2, label1), axis=1)
arr2.shape
```

```python
arr_full = np.concatenate((arr1, arr2), axis=0)
arr_full.shape
```
