```python
import matplotlib.pyplot as plt

def calculate_growth_time(initial_volume, target_volume, doubling_time):
    """
    Calculate the time it takes for a substance to reach a target volume
    given its initial volume and doubling time.
    
    Parameters:
        initial_volume (float): The initial volume of the substance.
        target_volume (float): The target volume to reach.
        doubling_time (float): The time it takes for the substance to double its volume.
        
    Returns:
        time (float): The time it takes to reach the target volume.
    """
    time = 0
    volume = initial_volume
    while volume < target_volume:
        volume *= 2
        time += doubling_time
    return time

def plot_growth(initial_volume, target_volume, doubling_time, interval=1):
    """
    Plot the growth of a substance over time until it reaches the target volume.
    
    Parameters:
        initial_volume (float): The initial volume of the substance.
        target_volume (float): The target volume to reach.
        doubling_time (float): The time it takes for the substance to double its volume.
        interval (float): The time interval for plotting points (default is 1 minute).
    """
    times = []
    volumes = []
    time = 0
    volume = initial_volume
    while volume < target_volume:
        times.append(time)
        volumes.append(volume)
        volume *= 2
        time += doubling_time
    plt.plot(times, volumes)
    plt.xlabel('Time (minutes)')
    plt.ylabel('Volume')
    plt.title('Volume Growth Over Time')
    plt.grid(True)
    plt.show()

# Constants
initial_volume_chestnut_bun = 1  # Assuming the volume of a single chestnut bun is 1 unit
target_volume_solar_system = 1e27  # Assumption for the volume of the solar system in cubic units
doubling_time = 5  # Doubling time in minutes

# Calculate growth time
growth_time = calculate_growth_time(initial_volume_chestnut_bun, target_volume_solar_system, doubling_time)
print("Time to cover the solar system with chestnut buns: {:.2f} minutes".format(growth_time))

# Plot growth
plot_growth(initial_volume_chestnut_bun, target_volume_solar_system, doubling_time)
```
