[Problem 5] Python functionalization

```python
# function 1
def quadric_function(x):
    return x**2
```

```python
# function 2
def exp_function(x):
    return 2*(x**2)+2**x
```

```python
# function 3
def sin_function(x):
    return np.sin(np.sqrt(x))
```

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):

    x = np.arange(x_range[0], x_range[1], x_range[2])
    y = function(x)

    x_diff = np.diff(x)
    y_diff = np.diff(y)
    gradient = y_diff/x_diff
    array_xy = np.vstack((x, y)).T

    
    return array_xy, gradient
```

```python
array_xy_quad, gradient_quad = compute_gradient(quadric_function)

plt.plot(gradient_quad)
plt.title("gradient of quadric function")
plt.show()
```

```python
array_xy_exp, gradient_exp = compute_gradient(exp_function)

plt.plot(gradient_exp)
plt.title("gradient of exp function")
plt.show()
```

```python
array_xy_sin, gradient_sin = compute_gradient(sin_function)

plt.plot(gradient_sin)
plt.title("gradient of sin function")
plt.show()
```

```python
print(np.min(array_xy_quad[:, 1]))
print(np.argmin(array_xy_quad[:, 1]))
```
